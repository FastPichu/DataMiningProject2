## Modeling

```python
X = df.drop(columns = ['Type'])
y = df['Type']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size = 0.2 ) #80% train, 20% test
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knnClf = KNeighborsClassifier(n_neighbors = 10)
```

```python
knnClf.fit(X_train, y_train)
```

```python
knnPredicted = knnClf.predict(X_test)
```

```python
knnPredicted
```

```python
knnClf.score(X_test, y_test)
```

```python
from sklearn.feature_selection import *
knnImp = SelectKBest(knnClf)
```

```python
knnImp
```

```python
from sklearn.metrics import *
```

```python
knnClfReport = classification_report(y_test, knnPredicted)
```

```python
print(knnClfReport)
```

```python
from sklearn.naive_bayes import CategoricalNB

nbClf = CategoricalNB(alpha = 1.0)
```

```python
nbClf.fit(X_train, y_train)
```

```python
nbPredicted = nbClf.predict(X_test)
```

```python
nbClfReport = classification_report(y_test, nbPredicted)
```

```python
print(nbClfReport)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfClf = RandomForestClassifier()
```

```python
rfClf.fit(X_train, y_train)

rfClf.score(X_test, y_test)
```

```python
rfClfPredicted = rfClf.predict(X_test)
```

```python
rfClfReport = classification_report(y_test, rfClfPredicted)
```

```python
print(rfClfReport)
```
